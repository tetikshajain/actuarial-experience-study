## Key Findings

- Overall mortality experience is higher than expected (A/E ≈ 1.09), indicating a modest underestimation of mortality risk.

- Male policyholders exhibit significantly higher mortality (A/E > 1.1), suggesting gender-based deviations from assumptions.

- Mortality deviations are concentrated in ages 50–70, where A/E ratios consistently exceed 1.0, indicating underestimation of risk in mid-to-late age segments.

- Claim costs are disproportionately driven by older age groups due to both higher mortality and larger exposure.

---

## Actuarial Interpretation

The observed deviations between actual and expected mortality suggest that current mortality assumptions may not fully capture risk in key demographic segments.

Segment-level analysis highlights the importance of refining assumptions based on age and gender to improve predictive accuracy.

---

## Recommendations

- Increase mortality assumptions for ages 50–70 by approximately **5–10%** to better reflect observed experience.

- Apply differentiated mortality assumptions by gender, particularly adjusting male mortality upward to reflect higher observed risk.

- Incorporate segment-level experience analysis into pricing and risk management processes to improve accuracy.

- Monitor mortality experience regularly to detect emerging trends and update assumptions accordingly.

---

## Business Impact

Refining mortality assumptions based on observed experience can:

- Improve pricing accuracy  
- Reduce unexpected claim volatility  
- Enhance risk management and capital allocation  
- Support more accurate underwriting decisions
