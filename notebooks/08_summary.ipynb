# Project Summary

This project was inspired by an interest in understanding how large datasets can be used to uncover meaningful patterns in real-world risks, particularly in areas where uncertainty plays a major role. Rather than focusing only on theoretical models, the goal was to explore how data-driven analysis can be used to evaluate and interpret real outcomes.

The project involved building a complete mortality analysis workflow using Canadian life table data. First, raw mortality data was cleaned and transformed into structured actuarial life tables, including key metrics such as mortality rates, survival probabilities, and life expectancy. A synthetic life insurance portfolio of 10,000 policies was then simulated to represent real-world exposure, incorporating attributes such as age, gender, policy duration, and coverage amounts.

Using this portfolio, an experience study was conducted by comparing observed mortality outcomes with expected values derived from mortality assumptions. This included calculating Actual-to-Expected (A/E) ratios at both the portfolio and segment level, as well as extending the analysis to financial impact through expected and actual claims.

Visualization techniques were used to highlight key trends, including deviations in mortality by age and gender, and the distribution of claim amounts. Finally, the results were translated into actionable insights and recommendations, demonstrating how data can be used to improve understanding of risk and inform decision-making.

## Key Achievements

Built an end-to-end analytical workflow from raw data to business insights
Simulated a realistic portfolio with over 10,000 policies and multi-million dollar exposure
Performed detailed Actual vs Expected analysis, identifying deviations in mortality across key segments
Translated statistical results into practical insights and recommendations
Integrated multiple tools including Python, SQL, and Excel to replicate real-world data workflows
